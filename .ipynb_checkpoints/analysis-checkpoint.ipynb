```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('orca_pc1_scaling.csv')
```

```python
fig, ax = plt.subplots(dpi=100)
FS = 16

ax.plot(df.nprocs, df.walltime, marker='.', color='black', mec='black', mfc='salmon', ms=20, ls='--', lw=2, mew=1.5)
ax.set_xlabel('Nprocs', fontsize=16)
ax.set_ylabel('Wall time [s]', fontsize=16)
ax.set_xticks([16, 32, 64, 128])
ax.set_xscale('log', base=2)
ax.tick_params(labelsize=FS-2)
ax.set_title('ORCA 5.0.2 timings: valinomycine @ HF/pc-1', fontsize=16)

ax.grid(ls=':', lw=0.75)
fig.tight_layout()
fig.savefig('orca_pc1_scaling.png', facecolor='white')
```
